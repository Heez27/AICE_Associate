```python
import pandas as pd
df = pd.read_csv('https://github.com/lovedlim/bigdata_analyst_cert_v2/blob/main/part1/ch3/type1_data1.csv?raw=true') # 깃허브 주소에 ?raw=true 붙이기
df
```

필터링, 최솟값, 중앙값

```python
# f5칼럼이 0이 아닌 행을 구하시오.
c1 = df['f5'] != 0
df = df[c1]
df
```

```python
# views 칼럼 결측치를 views칼럼의 최솟값으로 채우시오.
min = df['views'].min()
df['views'].fillna(min, inplace = True)
df
```

```python
# views 칼럼의 중앙값을 계산해 점수로 구하시오.
int(df['views'].median())
```

카테고리, 인덱스, 문자열 슬라이싱

```python
# subscribed 칼럼에서 가장 빈도수가 많은 날짜를 구하시오
values = df['subscribed'].value_counts()
values
```

```python
# 앞에서 구한 날짜의 일(day)값을 정수로 구하시오
int(values.index[0][-2:])
```

파생변수, 정렬, 인덱싱

```python
# 결측치가 있는 데이터(행)을 제거하기
df.dropna(inplace = True)
```

```python
# views칼럼을 f1칼럼으로 나눈 값을 새로운 칼럼으로 추가하기
df['views/f1'] = df['views']/df['f1']
df
```

```python
# 새로운 칼럼 중 가장 큰 값을 가진 행의 age를 정수로 구하기
df.sort_values('views/f1',ascending = False, inplace = True)
int(df.iloc[0,1])
```

값 변경, 정렬, 합계

```python
df['views'].fillna(0, inplace = True) # views칼럼의 데이터를 0으로 대체하기
df
```

```python
# views 칼럼에서 상위 10번째 값을 구하기
df = df.sort_values('views', ascending = False)
views_10thmin = df.iloc[:10]['views'].min()
views_10thmin
```

```python
# 상위 10개의 값을 상위 10번째 값으로 대체하기
df.iloc[:10,-1] = views_10thmin
df
```

```python
print(int(df['views'].sum())) # views 칼럼의 전체 합을 정수로 구하기
```

```python
# 다른 방법
df['views'].fillna(0, inplace = True)
views_10thmin = df['views'].nlargest(10).iloc[-1]
df.loc[df['views']>views_10thmin, 'views'] = views_10thmin
print(int(df['views'].sum()))
```

문자열 슬라이싱, 파생변수, 평균값

```python
# f4 칼럼 데이터에 FJ가 포하된 데이터 찾기
df['new'] = df['f4'].str[-2:]
cond = df['new'] == 'FJ'
FJ = df[cond]
FJ
```

```python
print(round(FJ['f2'].mean(),2)) # 찾은 데이터 중에서 f2 칼럼의 평균값 구하기
```

```python
# 다른 풀이
cond = df['f4'].str.contains('FJ')
print(round(df[cond]['f2'].mean(),2))
```

필터링, 분산

```python
# f3칼럼이 gold이면서 f2 칼럼이 2인 데이터 구하기
cond1 = df['f3']=='gold'
cond2 = df['f2']==2
df = df[cond1 & cond2]
df
```

```python
# 위 데이터에서 f1칼럼의 분산 구하기
print(round(df['f1'].var(),2))
```

값 변경(연산), 필터링 절대값

```python
df['age'] = df['age'] + 1
df
```

```python
cond1 = (df['age']>=30) & (df['age']<40)
cond2 = (df['age']>=20) & (df['age']<30)
print(round(abs(df[cond1]['views'].mean()-df[cond2]['views'].mean()),2))
```

이상치, IQR

```python
q1 = df['views'].quantile(.25)
q3 = df['views'].quantile(.75)
IQR = q3 - q1
```

```python
line1 = q1- 1.5 * IQR
line2 = q3+ 1.5 * IQR
cond1 = df['views']<line1
cond2 = df['views']>line2
df = df[cond1 | cond2]
print(sum(df['views']))
```
