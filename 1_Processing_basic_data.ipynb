```python
# drop으로 행 삭제하기
flight.drop(index = 0, axis = 0).head()
```

```python
flight.head() # 원본 데이터에서는 price 칼럼과 index=0인 행이 삭제안됨. inplace=True를 넣어야 원본에서 제거됨.
```

```python
# inplace = True로 원본데이터에서 삭제하기
flight.drop('price2', axis = 1, inplace=True)
flight.drop(index=0,axis=0,inplace = True)
flight.head()
```

```python
# 칼럼명 변경하기
flight.rename(columns = {'airline':'airline_name', 'source_city':'departure_city'},inplace=True)
flight.head()
```

```python
# 데이터프레임 정렬하기
flight.sort_values(by='Unnamed: 0', ascending = False, inplace = True)
flight
```

## **[데이터프레임 변형하기]**

```python
airline_group = flight.groupby('airline') # 그룹화까지만 함
airline_group.groups # 그룹 보여달라
```

```python
airline_group.count() # airline별 데이터 수 확인
```

```python
airline_group.min()
```

```python
airline_group.mean(numeric_only=True)
```

```python
airline_group.mean(numeric_only=True)[['price']]
```

```python
flight.groupby(['airline', 'arrival_time']).mean(numeric_only=True)
```

```python
flight.groupby(['airline', 'arrival_time']).mean(numeric_only=True).loc[[('AirAsia', 'Evening')]]
```

```python
flight.set_index(['airline','arrival_time']) # 인덱스로 그룹화하기
```

```python
flight.set_index(['airline', 'arrival_time']).groupby(level=[0]).mean(numeric_only=True)
```

```python
flight.set_index(['airline', 'arrival_time']).groupby(level=[0,1]).mean(numeric_only=True)
```

```python
# flight.set_index(['airline', 'arrival_time']).groupby(level=[0,1]).aggregate([np.mean, np.max])
flight.set_index(['airline', 'arrival_time']) \
      .select_dtypes(include='number') \
      .groupby(level=[0, 1]) \
      .aggregate([np.mean, np.max])
```
